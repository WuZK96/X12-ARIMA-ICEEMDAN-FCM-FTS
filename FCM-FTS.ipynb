```python
dataset_name = 'SH18-CR'

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=[10,5])

dataset1 = np.array(dat["CR"])[:144]
dataset_tdiff1 = tdiff.apply(dataset1)


ax[0].plot(dataset1)
ax[1].plot(dataset_tdiff1)

ax[0].set_title(dataset_name)
```

```python
partitioners = {}
partitioners_diff = {}


output = [ ]
test = dataset1[132:144]

for ss in range(1,101):

    partitioners_diff  = {}

    partitioner_diff  = FCM.FCMPartitioner(data=dataset1[:132], npart=64,  transformation=tdiff)
    partitioners_diff  = partitioner_diff 

    model2 = chen.ConventionalFTS(partitioner=partitioners_diff)
    model2.name=dataset_name
    model2.append_transformation(tdiff)
    model2.fit(dataset1[:132], save_model=True, file_path='model2'+dataset_name, order=1)


    model2 = cUtil.load_obj('model2'+dataset_name)
    forecasts = model2.predict(dataset1[131:143]) 
    
    output.append(forecasts)
    
ave = []
for j in range(0,12):
    sum = 0
    for i in range(0,100):
        sum += output[i][j]
    ave.append(sum/100)
    
print(ave)
rmse = np.sqrt(np.nanmean((test - ave) ** 2))
mae = np.nanmean(np.abs(test - ave))
mape = np.nanmean(np.abs(test - ave)/test)*100
print('RMSE=',rmse,'MAE=',mae,'MAPE=',mape)
np.savetxt('output.csv', ave, delimiter = ',')


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[20,10])
ax.plot(ave)
ax.plot(test)    
ax.set_title(dataset_name)
```
